```python
from faker import Faker # type: ignore
import random
import hashlib

fake = Faker('es_ES')

# Generar una lista de carreras universitarias
carreras_universitarias = [
    'Ingeniería Civil',
    'Medicina',
    'Derecho',
    'Arquitectura',
    'Economía',
    'Psicología',
    'Administración de Empresas',
    'Ingeniería Informática',
    'Periodismo',
    'Biología'
]

# Crear una función para generar RUTs válidos
def generar_rut():
    rut = str(random.randint(1000000, 25000000))
    rut += '-' + str(random.randint(0, 9))
    return rut

# Crear una función para generar contraseñas seguras
def generar_contraseña():
    caracteres = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789!@#$%^&*()_+=-"
    longitud = random.randint(8, 12)
    contraseña = ''.join(random.sample(caracteres, longitud))
    return contraseña

# Número de personas en la base de datos
num_personas = 100

# Crear la base de datos
base_de_datos = []

for _ in range(num_personas):
    nombre = fake.name()
    id_persona = _+1
    correo = fake.email()
    rut = generar_rut()
    contraseña = generar_contraseña()
    carrera = random.choice(carreras_universitarias)
    base_de_datos.append({
        'Nombre': nombre,
        'ID': id_persona,
        'Correo': correo,
        'RUT': rut,
        'Contraseña': contraseña,
        'Carrera Universitaria': carrera
    })

# Mostrar algunos registros de la base de datos
for i in range(5):
    print(base_de_datos[i])
```
